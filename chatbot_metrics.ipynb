```python
from nltk.translate.bleu_score import sentence_bleu
from textblob import TextBlob
from langdetect import detect

# Sample Inputs and Bot Responses
inputs = [
    "मेरे परिवार में मानसिक स्वास्थ्य के बारे में बात करना एक टैबू है।",
    "त्योहारों के दौरान मुझे हमेशा अपने परिवार के लिए खुश रहना पड़ता है, लेकिन अंदर से मैं तनाव में रहता हूँ।"
]

expected_responses = [
    "यह समझना मुश्किल है, लेकिन किसी प्रोफेशनल से बात करना फायदेमंद हो सकता है।",
    "आपके खुद के लिए समय निकालना महत्वपूर्ण है और अपने परिवार के साथ अपनी भावनाओं को साझा करना भी सहायक हो सकता है।"
]

bot_responses = [
    "Oh, I see! It can be challenging to open up to family members, especially when it comes to mental health. Have you considered talking to a professional?",
    "It's important to prioritize your own happiness while also taking care of your family's needs. Take time for yourself each day."
]

# Cultural Keywords for Sensitivity Check (Example for Hindi context)
cultural_keywords = {
    "Hindi": ["परिवार", "त्योहार", "टैबू", "प्रोफेशनल", "खुश रहना"]
}

# Function to Calculate BLEU Score
def calculate_bleu(reference, response):
    reference_tokens = [reference.split()]
    response_tokens = response.split()
    score = sentence_bleu(reference_tokens, response_tokens)
    return score

# Function for Sentiment Analysis
def analyze_sentiment(response):
    analysis = TextBlob(response)
    return analysis.sentiment.polarity

# Function for Language Detection
def detect_language(response):
    try:
        language = detect(response)
        return language
    except Exception as e:
        return str(e)

# Function for Cultural Sensitivity Check
def check_cultural_sensitivity(language, response):
    keywords = cultural_keywords.get(language, [])
    matched_keywords = [word for word in keywords if word in response]
    sensitivity_score = len(matched_keywords) / len(keywords) if keywords else 0
    return sensitivity_score, matched_keywords

# Evaluation
def evaluate_responses(inputs, expected_responses, bot_responses, language_code):
    results = []

    for i in range(len(inputs)):
        input_text = inputs[i]
        expected = expected_responses[i]
        bot_response = bot_responses[i]

        # BLEU Score
        bleu_score = calculate_bleu(expected, bot_response)

        # Sentiment Analysis
        sentiment = analyze_sentiment(bot_response)

        # Language Detection
        detected_language = detect_language(bot_response)

        # Cultural Sensitivity Check
        sensitivity_score, matched_keywords = check_cultural_sensitivity(language_code, bot_response)

        # Check if the response language matches the input language
        input_language = detect(input_text)
        language_match = (detected_language == input_language)

        results.append({
            "Input": input_text,
            "Expected Response": expected,
            "Bot Response": bot_response,
            "BLEU Score": round(bleu_score, 2),
            "Sentiment": round(sentiment, 2),
            "Detected Language": detected_language,
            "Language Match": language_match,
            "Cultural Sensitivity Score": round(sensitivity_score, 2),
            "Matched Cultural Keywords": matched_keywords
        })

    return results

# Run Evaluation for Hindi
results = evaluate_responses(inputs, expected_responses, bot_responses, "Hindi")

# Display Results
for result in results:
    print("Input:", result["Input"])
    print("Expected Response:", result["Expected Response"])
    print("Bot Response:", result["Bot Response"])
    print("BLEU Score:", result["BLEU Score"])
    print("Sentiment:", result["Sentiment"])
    print("Detected Language:", result["Detected Language"])
    print("Language Match:", result["Language Match"])
    print("Cultural Sensitivity Score:", result["Cultural Sensitivity Score"])
    print("Matched Cultural Keywords:", result["Matched Cultural Keywords"])
    print("-" * 50)
```

```python
# Import necessary libraries
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from transformers import pipeline
from langdetect import detect
import pandas as pd

# Example data
data = [
    {
        "input_language": "ar",
        "reference": "أشعر بالخجل عند التعبير عن مشاعري أمام عائلتي. كيف يمكنني التغلب على هذا الشعور؟",
        "response": "You're going through a tough time with a family member's mental health issues. It's okay to feel overwhelmed and alone, but know you're not. Many people understand what you're going through and there are resources available for support. Consider reaching out to a therapist or counselor for a safe space to talk and get guidance. Take things one step at a time and focus on self-care. Feel free to share more or ask questions, I'm here to listen and help."
    }
]

# Initialize sentiment analysis pipeline
sentiment_analyzer = pipeline("sentiment-analysis")

# Initialize smoothing function for BLEU
smoothing = SmoothingFunction().method1

# Function to calculate BLEU score
def calculate_bleu(reference, response):
    reference_tokens = reference.split()  # Tokenize reference
    response_tokens = response.split()  # Tokenize response
    return sentence_bleu([reference_tokens], response_tokens, smoothing_function=smoothing)

# Function to detect language match
def detect_language_match(input_language, response):
    detected_language = detect(response)
    return detected_language, detected_language == input_language

# Initialize results list
results = []

# Process each entry
for entry in data:
    reference = entry["reference"]
    response = entry["response"]
    input_language = entry["input_language"]

    # Calculate BLEU Score
    bleu_score = calculate_bleu(reference, response)

    # Perform Sentiment Analysis
    sentiment = sentiment_analyzer(response)[0]
    sentiment_score = sentiment["score"]
    sentiment_label = sentiment["label"]

    # Detect Language Match
    detected_language, language_match = detect_language_match(input_language, response)

    # Append results
    results.append({
        "Input Language": input_language,
        "Reference": reference,
        "Response": response,
        "BLEU Score": bleu_score,
        "Sentiment Score": sentiment_score,
        "Sentiment Label": sentiment_label,
        "Detected Language": detected_language,
        "Language Match": language_match
    })

# Create a DataFrame
results_df = pd.DataFrame(results)

# Display results
print(results_df)

# Optionally, save results to a CSV
results_df.to_csv("multilingual_response_metrics.csv", index=False)
```
